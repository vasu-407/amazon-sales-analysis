```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
print(df['Order Date'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Seaborn styling
sns.set(style="whitegrid")           # clean background
palette = sns.color_palette("pastel")  # use pastel color palette

# Prepare data
region_sales = df.groupby('Region')['Total Sales'].sum().reset_index()

# Plot using seaborn barplot
plt.figure(figsize=(8, 5))
sns.barplot(data=region_sales, x='Region', y='Total Sales', palette=palette)
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
category_sales = df.groupby('Product Category')['Total Sales'].sum()

plt.figure(figsize=(8, 5))
category_sales.plot(kind='bar', color='lightgreen')
plt.title('Total Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.show()
```

```python
plt.scatter(df['Unit Price'], df['Total Sales'])
plt.title("Scatter Plot: Unit Price vs Total Sales")
plt.xlabel("Unit Price")
plt.ylabel("Total Sales")
plt.tight_layout()
plt.show()
```

```python
plt.hist(df['Total Sales'])
plt.title("Histogram: Total Sales")
plt.xlabel("Total Sales")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
category_counts = df['Product Category'].value_counts()

plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Pie Chart: Product Categories")
plt.axis('equal')
plt.show()
```
